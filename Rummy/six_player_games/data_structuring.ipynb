Don't need to sort again, if we have the sorted dataset, or else sort

```python
df_sorted = df.sort_values(by="settlement_time").reset_index(drop=True) #sorting by settlement time
```

```python
df_sorted.head(18)  # Display the first few rows of the sorted dataframe
```

```python
# Assign player position (1 to 6) in each game
df_sorted['player_number'] = df_sorted.groupby("game_id").cumcount() + 1
```

```python
df_sorted.head(18)  # Display the first few rows with player numbers
```

```python
# extract the first occurrence of settlement_time per game_id and create a (game_id, settlement_time) map
settlement_order = df_sorted.drop_duplicates("game_id")[["game_id", "settlement_time"]]
```

```python
# Pivoting to get user_id_i columns (1 to 6)
pivoted_users = df_sorted.pivot(index="game_id", columns="player_number", values="user_id")
pivoted_users.columns = [f"user_id_{i}" for i in range(1, 7)]
pivoted_users = pivoted_users.reset_index() # reset game_id as a normal column
```

```python
#merging back settlement_time and sorting again using settlement_time
pivoted_users = pivoted_users.merge(settlement_order, on="game_id", how="left")
pivoted_users = pivoted_users.sort_values("settlement_time").reset_index(drop=True)
```

```python
pivoted_users.head(18)  # Display the first few rows of the pivoted dataframe
```

```python
# Extracting the winner's information, assuming win_position == 1 indicates the winner
game_meta = df_sorted[df_sorted["win_position"] == 1].drop_duplicates(subset=["game_id"])[
    ["game_id", "user_id", "s_date", "pkr"]
].rename(columns={"user_id": "winner_id"})
```

```python
game_meta.head(18)  # Display the first few rows of the winner's information per game
```

```python
final_df = pivoted_users.merge(game_meta, on="game_id", how="left")
```

```python
final_df = final_df.sort_values("settlement_time").reset_index(drop=True) #sorting by settlement time again just to be sure
final_df.head(18) 
```

```python
# Initialize a game count dictionary
user_game_counts = dict()

# initialize game_count columns for each user_id
for i in range(1, 7):
    final_df[f"game_count_{i}"] = 0

# iterating through each game row and fill in the game_count for each user_id
for idx, row in final_df.iterrows(): 
    for i in range(1, 7): 
        uid = row[f"user_id_{i}"] # get the user_id for player i
        if pd.isna(uid):
            continue  # skip if user_id is missing
        if uid not in user_game_counts: # if the user_id is not in the dictionary, initialize it at 1
            user_game_counts[uid] = 1
        else:
            user_game_counts[uid] += 1 # if it is already in the dictionary, increase the count by 1
        # set the game_count for player i   
        final_df.at[idx, f"game_count_{i}"] = user_game_counts[uid] # set the game_count for player i
```
