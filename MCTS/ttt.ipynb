```python
import pandas as pd

data = pd.read_csv("mcts_vs_deterministic_results2.csv")
sec_player = data[data['Deterministic_Player']=='X']
print(sec_player.head())
firs_player = data[data['Deterministic_Player']=='O']
print(firs_player.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate loss %
sec_player['loss %'] = sec_player['Wins_X'] * 100  # scale to percent if not already

# Set a better style
sns.set(style='whitegrid', context='notebook', palette='muted')

# Create the plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=sec_player, x='MCTS_Iterations', y='loss %', marker='o', linewidth=2)

# Improve titles and labels
plt.title('Loss % of MCTS strategy as 2nd player', fontsize=14, fontweight='bold')
plt.xlabel('MCTS Iterations', fontsize=12)
plt.ylabel('Loss Percentage (%)', fontsize=12)

# Enhance readability
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate loss %
firs_player['loss %'] = firs_player['Wins_O'] * 100  # scale to percent if not already

# Set style
sns.set(style='whitegrid', context='notebook', palette='muted')

# Create the plot
plt.figure(figsize=(10, 6))
ax = sns.lineplot(data=firs_player, x='MCTS_Iterations', y='loss %', marker='o', linewidth=2)

# Annotate each point with rounded value
for x, y in zip(firs_player['MCTS_Iterations'], firs_player['loss %']):
    plt.text(x+40, y + 1, f"{y:.2f}%", ha='center', va='bottom', fontsize=12, color='black')

# Titles and labels
plt.title('Loss % of MCTS strategy as 1st player', fontsize=25, fontweight='bold')
plt.xlabel('MCTS Iterations', fontsize=20)
plt.ylabel('Loss Percentage (%)', fontsize=20)
plt.ylim(-5,100)

# Aesthetics
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```
